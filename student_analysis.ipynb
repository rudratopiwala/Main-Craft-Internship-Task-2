### Task 1

```python
# load dataset
import pandas as pd

data=pd.read_csv(r"D:\jupyter practice\Datasets\student-mat.csv",sep=";")
data.head(3)
```

### Task 2 : Explore & Clean data

### This step helps identify if the dataset contains any null or missing values.

```python
# check missing values
data.isnull().sum()
```

```python
#remove duplicates
print("Data with duplication :",data.duplicated().sum())
```

### This step removes any duplicate rows from the dataset.

```python
data.drop_duplicates()
print("Data after remove duplictes :",data.duplicated().sum())
```

```python
#inspects Dataset shape
data.shape
```

```python
#inspect dataset dtypes
data.info()
```

### Task 3: Analysis Questions

```python
#Average final grade (G3). 
print("Average of final grade (G3) : ",data["G3"].mean())
```

```python
#How many students scored above 15? 
print("total students scored above 15 : ",data[data['G3']>15].shape[0])
```

```python
#Is study time correlated with performance?
print("Correlation between study time and final grade :",data['studytime'].corr(data['G3']))
```

```python
#Which gender performs better on average?
print("gender who performs better:",data.groupby('sex')['G3'].mean())
```

### Task 4 : Visualizations

```python
# Histogram of grades.
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,4))
plt.hist(data['G3'],edgecolor="black")
plt.title("Distribution of final grades")
plt.xlabel("FInal grade(G3)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatterplot: study time vs grades.
plt.figure(figsize=(7,4))
plt.scatter(data['studytime'],data['G3'])
plt.title("study time vs final grade")
plt.xlabel("study time")
plt.ylabel("final garde")
plt.show()
```

```python
#Bar chart: male vs female average score. 
plt.figure(figsize=(7,4))
g=data.groupby('sex')['G3'].mean()
g.plot(kind="bar")
plt.title("Avg final grade by gender")
plt.xlabel("Gender")
plt.ylabel("Avg G3 score")
plt.show()
```
